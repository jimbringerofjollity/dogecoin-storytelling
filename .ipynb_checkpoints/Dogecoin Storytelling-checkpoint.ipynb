# Dogecoin Storytelling

**Introduction**

During January 2021, dogecoin mooned. In this notebook, we tell the story through an exploratory data analysis. Following a guide with directions on how to import cryptocurrency data from Binance (https://bit.ly/2MPMqkJ), we import the necessary python packages and adapt the method `get_bars` to allow the specification of start and end times and limits on the number of points.

```python
import requests        # for making http requests to binance
import json            # for parsing what binance sends back to us
import pandas as pd    # for storing and manipulating the data we get back
import numpy as np     # numerical python, i usually need this somewhere 
                       # and so i import by habit nowadays

import matplotlib.pyplot as plt # for charts and such
    
from datetime import datetime as dt  # for dealing with times
```

```python
def get_bars(symbol = 'BTCUSDT', interval = '1d', start = '01.01.2020 00:00:00,00', end = '01.02.2021 00:00:00,00', limit=500):
   startTime = int(dt.strptime(start,
                           '%d.%m.%Y %H:%M:%S,%f').timestamp() * 1000)
   endTime = int(dt.strptime(end,
                           '%d.%m.%Y %H:%M:%S,%f').timestamp() * 1000)
   root_url = 'https://api.binance.com/api/v3/klines'
   url = root_url + '?symbol=' + symbol + '&interval=' + interval + '&startTime=' + str(startTime) + '&endTime=' + str(endTime) + '&limit=' + str(limit)
   data = json.loads(requests.get(url).text)
   df = pd.DataFrame(data)
   df.columns = ['open_time',
                 'o', 'h', 'l', 'c', 'v',
                 'close_time', 'qav', 'num_trades',
                 'taker_base_vol', 'taker_quote_vol', 'ignore']
   df.index = [dt.fromtimestamp(x/1000.0) for x in df.close_time]
   return df
```

First we test our method on the bitcoin price from January 2020 to January 2021. As you can see below, the price of bitcoin went down during the initial week of the COVID-19 lockdown in March but steadily went up as the Federal Reserve started printing money. Towards the end of 2020, bitcoin mooned.

```python
btcusdt = get_bars()
btcusdt = btcusdt['c'].astype('float')
btcusdt.plot()
plt.title('Bitcoin Price from January 2020 to January 2021')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

A similar story can be told for dogecoin, although its dramatic increase was in January 2021. As you can see from the graph below, the price of dogecoin stayed relatively flat for most of 2020, but in January 2021 you see two sudden spikes upward in the dogecoin price, a smaller one around the beginning of January 2021 and a larger one towards the end of January 2021.

```python
dogeusdt = get_bars('DOGEUSDT')
dogeusdt = dogeusdt['c'].astype('float')
dogeusdt.plot()
plt.title('Dogecoin Price from January 2020 to January 2021')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

Comparing January 2020 and January 2021, you can see that January 2020 looks more like a normal month, whereas January 2021 looks more anomalous, with sudden spikes upward in the price.

```python
dogeusdt_interval1 = get_bars('DOGEUSDT', interval='1h', start = '01.01.2020 00:00:00,00', end = '01.02.2020 00:00:00,00',limit=1000)
dogeusdt_interval1 = dogeusdt_interval1['c'].astype('float')
dogeusdt_interval1.plot()
plt.title('Dogecoin Price in January 2020')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

```python
dogeusdt_interval2 = get_bars('DOGEUSDT', interval='1h', start = '01.01.2021 00:00:00,00', end = '01.02.2021 00:00:00,00',limit=1000)
dogeusdt_interval2 = dogeusdt_interval2['c'].astype('float')
dogeusdt_interval2.plot()
plt.title('Dogecoin Price in January 2021')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```
